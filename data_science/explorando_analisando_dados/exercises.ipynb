```python
import pandas as pd
```

```python
# impontando base de dados
df_1 = pd.read_csv("arquivos/serenatto_2sem_2023.csv", sep=",")
df_1
```

```python
# Somantório de todas as vendas do produto "Ratatouille" do período
df_1.query("produto == 'Ratatouille'")["valor"].sum()

# Frequência dos métodos de pagamento para o produto "Ratatouille"
df_1.query("produto == 'Ratatouille'")["metodo_pagamento"].value_counts()

# dados referentes ao produto "Tiramisù"
df_1.query("produto == 'Tiramisù'")

# linhas em que o valor do produto é MAIOR que 10 e o método de pagamento é qualquer um, EXCETO DINHEIRO
df_1.query("valor > 10 and metodo_pagamento != 'Dinheiro'")
```
